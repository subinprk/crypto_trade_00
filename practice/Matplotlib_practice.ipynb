```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = x**2 + 1

plt.figure()
plt.plot(x, y)
plt.show()
```

```python
z = -x ** 2 + 2

plt.figure()
plt.figure(figsize=(16, 10))
plt.title("my title")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x, y, label='y')
plt.plot(x, z, label='z')
plt.legend()
plt.show()
```

```python
plt.subplot(221)
plt.plot(x,y)
plt.show()

plt.subplot(122)
plt.plot(x,z)
plt.show()

plt.subplot(223)
plt.plot(x,z)
plt.show()
```
